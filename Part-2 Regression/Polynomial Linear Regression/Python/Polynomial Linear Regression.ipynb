Data Preprocessing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

dataset = pd.read_csv('Position_Salaries.csv')
# we only take second column i.e Level which corresponds to the Position
X = dataset.iloc[:, -2].values
y = dataset.iloc[:, -1].values

# we do not split into training and test set since data is less and we want to include most of it for training
# we will the model later to predict salary for a random level value
```

---

Training the (Simple) Linear Regression Model on whole set

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X.reshape(len(X), 1), y.reshape(len(y), 1))
```

---

Training the Polynomial Linear Regression Model on whole set

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X.reshape(len(X), 1))
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y.reshape(len(y), 1))
```

---

Visualizing the Linear Regression results

```python
plt.scatter(X, y, color='red')
plt.plot(X, linear_reg.predict(X.reshape(len(X), 1)), color='blue')
plt.title('Salary vs Level- Linear Regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

---

Visualizing the Polynomial Regression results

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly), color='blue')
plt.title('Salary vs Level- Polynomial Regression (D=2)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

---

Retraining the Polynomial Regression model with higher power

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg_2 = PolynomialFeatures(degree=4)
X_poly_2 = poly_reg_2.fit_transform(X.reshape(len(X), 1))
lin_reg_22 = LinearRegression()
lin_reg_22.fit(X_poly_2, y.reshape(len(y), 1))
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_22.predict(X_poly_2), color='blue')
plt.title('Salary vs Level- Polynomial Regression (D=4)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

---

Predicting a new result with Linear Regression

```python
lin_reg_pred = lin_reg.predict([[6.5]])
print(lin_reg_pred)

#salary is over predicted, way too much
```

Predicting a new result with Polynomial Regression

```python
#we need to input the transformed value with an input as [[6.5]]

lin_reg2_pred = lin_reg_22.predict(poly_reg_2.fit_transform([[6.5]]))
print(lin_reg2_pred)
```
